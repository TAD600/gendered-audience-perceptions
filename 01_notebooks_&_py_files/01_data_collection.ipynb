##### Data Collection steps:
1. Logging into https://console.cloud.google.com/ and making a new project
2. Enabling YouTube Data API v3 for that project and creating credentials (Multiple API keys does not increases daily quota. Hence, created one key per project)
3. I fetched YouTube Channel ID's from this free online platform `https://commentpicker.com/youtube-channel-id.php`
4. Downloading videos for all five artistic category and making five separate corpuses for the categories
5. Making 'gender' and 'art' column in each corpus
6. Filling up the 'gender' columns for personal channels of artists. 
7. Saving all five dataframes in the data folder as csv files
8. Parsing comment date column into two new columns 'date', and 'year'
9. Removing the rows where the 'Comment' column is empty
10. Concatenating all five dataframes
11. Separating it into two based on gender handcoding needed or not
12. Saving two separate csv files
